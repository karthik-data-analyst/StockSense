```python
from groq import Groq
import pandas as pd
from datetime import datetime

# ---- YOUR FREE API KEY ----
API_KEY = ""

def run_stocksense():

    # ---- HEADER ----
    print("\n")
    print("=" * 40)
    print("   STOCKSENSE - Daily Restock Report")
    print(f"   {datetime.today().strftime('%d %B %Y')}")
    print("=" * 40)
    print("Analyzing your sales data...")
    print("\n")

    # ---- READ EXCEL ----
    df =  pd.read_excel(r"D:\Data Analytics\Data Analyst Project\SalesSense\salessense_data.xlsx")

    # ---- SHOW DATA ----
    print("Today's Sales Summary:")
    print("-" * 40)
    print(df.to_string(index=False))
    print("\n")

    # ---- SEND TO GROQ AI ----
    client = Groq(api_key=API_KEY)

    response = client.chat.completions.create(
        model="llama3-70b-8192",
        messages=[
            {
                "role": "system",
                "content": """
You are StockSense, an AI restock assistant 
for small shops in Tamil Nadu.

Rules:
- Maximum 5 lines only
- Say exactly HOW MANY units to restock
- Use emojis for easy reading
- End with one priority action for today
- Be direct, owner is busy
"""
            },
            {
                "role": "user",
                "content": f"Here is today's shop data:\n{df.to_string()}\n\nWhat should I restock today?"
            }
        ]
    )

    # ---- SHOW RESULT ----
    print("StockSense Recommendation:")
    print("-" * 40)
    print(response.choices[0].message.content)
    print("\n")
    print("=" * 40)
    print("   Powered by StockSense AI")
    print("=" * 40)
    print("\n")

# ---- RUN ----
run_stocksense()
```
